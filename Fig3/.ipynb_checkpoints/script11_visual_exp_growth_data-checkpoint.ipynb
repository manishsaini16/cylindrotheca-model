## error calculation and tables preparation

```python
df_gr_expt.head()
```

```python
df_gr_expt_mean = df_gr_expt.groupby(['sample_type']).mean().reset_index()
df_gr_expt_sd= df_gr_expt.groupby(['sample_type']).std().reset_index()
df_gr_expt_mean_sd = pd.merge(df_gr_expt_mean, df_gr_expt_sd, on='sample_type', how='left')
df_gr_expt_mean_sd.columns = ['sample_type', 'growth_rate_mean', 'growth_rate_sd']
df_gr_expt_insili = pd.merge(df_gr_expt_mean_sd, df_gr_insili, on='sample_type', how='left')
df_gr_expt_insili.drop(labels='data_type', axis=1, inplace = True)
df_gr_expt_insili.rename(columns= {'growth_rate': 'growth_rate_insilico'}, inplace= True)
df_gr_expt_insili.head()
```

```python
error = ((df_gr_expt_insili.growth_rate_mean.sub(df_gr_expt_insili.growth_rate_insilico))/df_gr_expt_insili.growth_rate_mean)*100
error = error.abs()
```

```python
df_gr_expt_insili['error_rate'] = error
```

```python
df_gr_expt_insili
```

```python
df_gr_expt_insili.columns = ['Growth conditions', 'Growth rate (mean)(Experimental)', 'Growth rate (S.D.)', 'Growth rate (In silico)', 'Error rate']
```
